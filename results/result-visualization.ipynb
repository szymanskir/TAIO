```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Eksperymentalne wyznaczenie złożoności

## Warunek V

```python
exact_data_vertices = pd.read_csv('exact-vertex-complexity-estimation.csv')
results = plt.scatter(x=exact_data_vertices.index, y=exact_data_vertices['czas obliczeń'], label='Algorytm')
k_1 = 0.0002
k_2 = 0.00002
x_linspace = np.arange(1, 18, 0.1)
upper_bound = plt.plot(x_linspace, k_1 * 8 * np.power(2, x_linspace), ':',label='Górne ograniczenie')
lower_bound = plt.plot(x_linspace, k_2 * 8 *np.power(2, x_linspace), '--', label='Dolne ograniczenie')
plt.ylim((0, 5))
plt.grid()
plt.xlabel('Liczba wierzchołków w zmiennym grafie wejściowym')
plt.ylabel('Czas obliczeń')
plt.legend()
```

W celu oszacowania złożoności algorytmu dokładnego przeprowadzono symulacje dla losowych par grafów, gdzie jeden miał ustaloną liczb wierzchołków - 8, zaś drugi przyjmował liczbę wierzchołków z przedziału $[1, 14]$. Symulację powtórzono dziesięć razy i dla każdego rozmiaru drugiego grafu obliczono średnią z otrzymanych czasów.

Zgodnie z rozważaniami teoretycznymi algorytm powinien mieć złożoność rzędu $\bigO\left(2^{|V_1||V_2|}\right)$. W rozważanym przypadku jeden z grafów ma stałą liczbę wierzchołków 8, zatem złożoność powinna być rzędy $\bigO\left(2^{|V_2|}\right)$. W celu znalezienia stałych

<!--Definicja ograniczenai dolnego i górnego-->

Na wykres naniesiono wykresy funkcji $0.0002 \cdot 2^{|V_2|}$ oraz $0.00002 \cdot 2^{|V_2|}$. Znaleziono takie stałe $0.0002$ oraz $0.00002$, które dla każdengo $n \geq 9$ spełniają powyższą definicję.
